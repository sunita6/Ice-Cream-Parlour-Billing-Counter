```python
class Queue:
    def __init__(self):
        self.que=[]
        self.front=-999
        self.rear=-999
        
    def isEmpty(self):
        return self.que==[]
    
    def Enqueue(self,data):
        self.que.append(data)
        if self.rear==-999:
            self.front=0
            self.rear=0
        else:
            self.rear=self.rear+1
    
    def Dequeue(self):
        if self.front==-999:
            return False
        elif self.front==self.rear:
            self.front=self.rear=-999
            return self.que.pop(0)
        else:
            self.rear=self.rear-1
            return self.que.pop(0)
        
    def Display(self):
        self.rear=len(self.que)-1
        if self.front==self.rear:
            print(self.que[0],"<-----------------Order getting prepared")
        else:
            print(self.que[self.front],"<---------------Order getting prepared")
            for x in range(self.front+1,self.rear):
                print(self.que[x])
            print(self.que[self.rear],"<----------------Last order in waiting list")
        print("********************************************************************\n")
        
q=Queue()   

```
